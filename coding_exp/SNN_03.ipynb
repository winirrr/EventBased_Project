```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
tau_m = 10 #ms
V_rest = -65 #mV
V_th = -56 #mV
V_reset = -65 #mV
R = 10.0  # M
I = 1.0  # nA

# Stimulate settings
dt = 0.1 #ms
T  = 100 #ms
time = np.arange(0, T, dt)
V = np.zeros_like(time)
V[0] = V_rest

# Stimulating loop
for t in range(1, len(time)):
    dV = (-(V[t-1] - V_rest) + R*I) * (dt/ tau_m)
    V[t] = V[t-1] + dV
    if V[t] >= V_th:
        V[t] = V_reset # Spike and reset

# Plotting
plt.figure(figsize=(10, 4))
plt.plot(time, V, label='Membrane Potential')
plt.axhline(V_th, color='r', linestyle='--', label='Threshold')
plt.axhline(V_rest, color='b', linestyle='--', label='Resting Potential')
plt.xlabel('Time (ms)')
plt.ylabel('Membrane Potential (mV)')
plt.title('Leaky Integrate-and-Fire Neuron Simulation')
plt.legend()
plt.show()
```
